```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.mode.chained_assignment = None # disable SettingWithCopyWarning

track_df = pd.read_csv("Track_List.csv") #get data frame for tracks
track_df.columns #get attributes for tracks
```

```python
track_df.head()
```

```python
track_df.info()
```

```python
track_df.describe()
```

```python
track_df['Play-to-Skip Ratio'] = track_df['Plays'] / track_df['Skips']
track_df.head(10)
```

```python
track_df.describe()
```

```python
sns.distplot(track_df['Play-to-Skip Ratio'])
```

```python
from scipy.stats import skew
skew(track_df['Play-to-Skip Ratio'])
```

```python
sns.countplot(track_df['Skips'])
```

```python
sns.set(rc={'figure.figsize':(6,6)})
sns.heatmap(track_df.corr(), annot=True, fmt='.2f', linewidth=2)
```

```python
#songs I skip (within first 20 sec) more than I play in full
least_liked = track_df.loc[track_df['Play-to-Skip Ratio'] < 1] 
least_liked.head(10)
```

```python
#songs I play 20+ times more than I skip (within first 20 sec)
most_liked = track_df.loc[track_df['Play-to-Skip Ratio'] >= 20] 
most_liked.head(10)
```

```python
#Let's see how many artists are among my least liked songs
plt.figure(figsize=(12,8)) # this creates a figure 12 inch wide, 8 inch high
ax = sns.countplot(x="Artist", data=least_liked)

ax.set_xticklabels(ax.get_xticklabels(), rotation=50, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Now let's see how many artists are among my most liked songs
plt.figure(figsize=(12,8))
ax = sns.countplot(x="Artist", data= most_liked)

ax.set_xticklabels(ax.get_xticklabels(), rotation=50, ha="right")
plt.tight_layout()
plt.show()
```

```python
# want to see how many songs least liked are apart of The Wall or Dark Side of the Moon
# need to keep those since they are apart of the overall experience
# I don't have an album column; might have to add that in at some point
floyd_least_liked = least_liked.loc[least_liked['Artist'] == 'Pink Floyd']
floyd_least_liked.head(len(floyd_least_liked)) # show whole list instead of first few rows
```

```python
# remove all Pink Floyd songs from least_liked except for 'Dogs' and 'Welcome to the Machine'
indexes = least_liked[(least_liked['Artist'] == 'Pink Floyd') & (least_liked['Title'] != 'Dogs') & (least_liked['Title'] != 'Welcome to the Machine')].index
least_liked.drop(indexes , inplace=True)
least_liked.info()
```
